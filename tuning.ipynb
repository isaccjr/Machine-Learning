```python
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Bank_Personal_Loan_Modelling.xlsx')
```

```python
df.head()
```

```python
df.drop('ID', axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Personal_Loan', axis=1), df['Personal_Loan'], test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import accuracy_score
```

```python
modelo = {'DecisionTreeClassifier': {'modelo':DecisionTreeClassifier(),
                                     'parametros':{'max_depth': [1, 5, 10, 20],
                                                   'min_samples_split': [2, 5, 10],
                                                   'min_samples_leaf': [1, 5, 10]} },
          'RandomForestClassifier': {'modelo':RandomForestClassifier(),
                                     'parametros':{'n_estimators': [10, 50, 100, 200],
                                                   'max_depth': [5, 10, 20],
                                                   'min_samples_split': [2, 5, 10],
                                                   'min_samples_leaf': [1, 5, 10]}}}
```

```python
for model in modelo.keys():
    untunned = modelo[model]['modelo']
    untunned.fit(X_train, y_train)
    y_pred_train = untunned.predict(X_train)
    y_pred_test = untunned.predict(X_test)
    print(model)
    print('Untunned Train Accuracy: ',accuracy_score(y_train, y_pred_train))
    print('Untunned Test Accuracy: ',accuracy_score(y_test, y_pred_test))
    print(untunned.get_params())
    cv = GridSearchCV(modelo[model]['modelo'], modelo[model]['parametros'], cv=5 , scoring='accuracy')
    cv.fit(X_train, y_train )
    tunned = cv.best_estimator_
    y_pred_train = tunned.predict(X_train)
    y_pred_test = tunned.predict(X_test)
    print(model)
    print('Tunned Train Accuracy: ',accuracy_score(y_train, y_pred_train))
    print('Tunned Test Accuracy: ',accuracy_score(y_test, y_pred_test))
    print(cv.best_params_)
```
