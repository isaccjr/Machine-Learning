# Cálculo do RMF

```python
df_RFM = df.groupby(['CustomerID', 'InvoiceNo','InvoiceDate'])['billing'].sum()
df_RFM
```

```python
#Cria um novo DataFrame com as colunas necessárias
# Create a new DataFrame with the required columns
df_rfm = pd.DataFrame({
    'InvoiceDate': df_RFM.index.get_level_values('InvoiceDate'),
    'InvoiceNo': df_RFM.index.get_level_values('InvoiceNo'),
    'TotalBilling': df_RFM.values,
    'CustomerID': df_RFM.index.get_level_values('CustomerID')
})
# Organiza pela Data da Nota Fiscal
# Sort the DataFrame by InvoiceDate
df_rfm = df_rfm.sort_values(by='InvoiceDate')

# Calcula a recência com a diferença entre o lançamento da última nota fiscal do sistema e a última pertecente ao cliente
# Calculate the Recency as the difference between the last invoice date and the last invoice date of the given Custumer
df_rfm['Recency'] = (last_date - df_rfm['InvoiceDate']).dt.days

# Calcula o valor total vendido para cada cliente
# Calculate the Monetary value as the sum of TotalBilling for each CustomerID
df_rfm['Monetary'] = df_rfm.groupby('CustomerID')['TotalBilling'].transform('mean')

# Calcula a frequência como a quantidade de notas fiscais de cada cliente 'CustumerID'
# Calculate the Frequency as the count of invoices for each CustomerID
df_rfm['Frequency'] = df_rfm.groupby('CustomerID')['InvoiceNo'].transform('count')

# Seleciona as colunas
# Select the desired columns
df_rfm = df_rfm[['CustomerID', 'Recency', 'Monetary', 'Frequency']]

# Mostra o DataFrame
# Print the DataFrame
print(df_rfm)
```

# Gerando o pipeline

```python
pipeline_df = pd.read_csv('data.csv')
pipeline_df.head()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
import pandas as pd

# Custom transformer for dropping missing values
class DropMissingValues(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return X.dropna()

# Custom transformer for dropping incorrect prices and quantities
class DropIncorrectValues(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        condition = (X['UnitPrice'] <= 0) | (X['Quantity'] <= 0)
        return X.drop(X[condition].index)

# Custom transformer for converting InvoiceDate to datetime and filtering
class ConvertDateAndFilter(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        self.last_date = X['InvoiceDate'].max()
        return self

    def transform(self, X):
        X['InvoiceDate'] = pd.to_datetime(X['InvoiceDate'])
        condition = (X['UnitPrice'] > 5000) | (X['Quantity'] > 10000)
        return X.drop(X[condition].index)

# Custom transformer for calculating billing
class CalculateBilling(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X['billing'] = X['UnitPrice'] * X['Quantity']
        return X

# Custom transformer for calculating RFM features
class CalculateRFM(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        self.last_date = X['InvoiceDate'].max()
        return self

    def transform(self, X):
        df_RFM = X.groupby(['CustomerID', 'InvoiceNo','InvoiceDate'])['billing'].sum()
        df_rfm = pd.DataFrame({
            'InvoiceDate': df_RFM.index.get_level_values('InvoiceDate'),
            'InvoiceNo': df_RFM.index.get_level_values('InvoiceNo'),
            'TotalBilling': df_RFM.values,
            'CustomerID': df_RFM.index.get_level_values('CustomerID')
        })
        df_rfm = df_rfm.sort_values(by='InvoiceDate')
        df_rfm['Recency'] = (self.last_date - df_rfm['InvoiceDate']).dt.days
        df_rfm['Monetary'] = df_rfm.groupby('CustomerID')['TotalBilling'].transform('mean')
        df_rfm['Frequency'] = df_rfm.groupby('CustomerID')['InvoiceNo'].transform('count')
        return df_rfm[['CustomerID', 'Recency', 'Monetary', 'Frequency']].drop_duplicates()

# Creating the pipeline
data_preprocessing_pipeline = Pipeline(steps=[
    ('drop_missing_values', DropMissingValues()),
    ('drop_incorrect_values', DropIncorrectValues()),
    ('convert_date_and_filter', ConvertDateAndFilter()),
    ('calculate_billing', CalculateBilling()),
    ('calculate_RFM', CalculateRFM())
])

# Now you can fit and transform your dataframe with the pipeline
df_transformed = data_preprocessing_pipeline.fit_transform(pipeline_df)
```
